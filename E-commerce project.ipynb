# Superstore Dataset cleaning and visualization project


```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import plotly.colors as color
pio.templates.default = 'plotly_white'
```

```python
da = pd.read_csv(r"C:\Users\dell\Downloads\Sample - Superstore.csv",encoding='latin1',)
da
```

```python
da.head()
```

```python
da.info()
```

```python
da.describe()
```

# converting date column 

```python
da['Order Date'] = pd.to_datetime(da['Order Date'], errors='coerce')
da['Ship Date'] = pd.to_datetime(da['Ship Date'])
da.info()
```

# Create three column for monthly, year and weekly analysis

```python
da['Order Month'] = da['Order Date'].dt.month
da['Order Year'] = da ['Order Date'].dt.year
da['Order Day of Week'] = da['Order Date'].dt.day_of_week
da
```

# Monthly sales analysis

```python
Sales_by_month = da.groupby('Order Month')['Sales'].sum().reset_index()
Sales_by_month
```

```python
fig = px.line(Sales_by_month,
                x = 'Order Month',
                y='Sales',
                 title= 'Monthly Sales Analysis')

fig.write_image("monthly_sales.png")
from IPython.display import Image
Image("monthly_sales.png")
```

# Sales by category

```python
Sales_by_Category = da.groupby('Category')['Sales'].sum().reset_index()
Sales_by_Category
```

```python
fig = px.pie (Sales_by_Category,
             values = 'Sales',
             names ='Category',
             hole = 0.3, 
             color_discrete_sequence=px.colors.qualitative.Pastel )
fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title_text = 'Sales Analysis Category', title_font = dict(size = 24))

fig.write_image("sales_by_category.png")
from IPython.display import Image
Image("sales_by_category.png")
```

# Sales by Sub-Category

```python
Sales_by_SubCategory = da.groupby('Sub-Category')['Sales'].sum().reset_index()
Sales_by_SubCategory
```

```python
fig =px.bar(
     data_frame=Sales_by_SubCategory,
     x= 'Sub-Category',
     y= 'Sales',
     title='Sales Analysis Sub-Category'
)

fig.write_image("sales_by_subcategory.png")
from IPython.display import Image
Image("sales_by_subcategory.png")
```

# Monthly profit analysis

```python
Profit_by_Month = da.groupby('Order Month')['Profit'].sum().reset_index()
Profit_by_Month
```

```python
fig =px.line(
    Profit_by_Month,
    x='Order Month',
    y='Profit',
    title='Monthly Profit Analysis '
)

fig.write_image("profit_by_month.png")
from IPython.display import Image
Image("profit_by_month.png")
```

# Profit by Category 


```python
Profit_by_Category = da.groupby('Category')['Profit'].sum().reset_index()
Profit_by_Category
```

```python
fig=px.pie(data_frame=Profit_by_Category,
           values='Profit',
           names='Category',
           hole=0.5,
           color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text = 'Profit Analysis by Category', title_font = dict(size=24))

fig.write_image("profit_by_category.png")
from IPython.display import Image
Image("profit_by_category.png")
```

# Profit by Sub-Category

```python
Profit_by_SubCategory = da.groupby('Sub-Category')['Profit'].sum().reset_index()
Profit_by_SubCategory
```

```python
fig = px.bar(
         Profit_by_SubCategory,
         x='Sub-Category',
         y='Profit',
         title='Profit analysis in Sub-Category'  
)

fig.write_image("profit_by_subcategory.png")
from IPython.display import Image
Image("profit_by_subcategory.png")
```

# Sales and Profit by Customer Segment

```python
Sales_and_Profit_by_Segment = da.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
Sales_and_Profit_by_Segment
```

```python
color_palette = color.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x=Sales_and_Profit_by_Segment['Segment'],
                     y=Sales_and_Profit_by_Segment['Sales'],
                    name='Sales',
                    marker_color = color_palette[0]))
fig.add_trace(go.Bar(x=Sales_and_Profit_by_Segment['Segment'],
                     y=Sales_and_Profit_by_Segment['Profit'],
                     name='profit',
                     marker_color= color_palette[1]))
fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title= 'Customer Segment', yaxis_title = 'Amount')

fig.write_image("sales_and_profit_by_segment.png")
from IPython.display import Image
Image("sales_and_profit_by_segment.png")
```

# Sales to Profit ratio

```python
Sales_and_Profit_by_Segment = da.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
Sales_and_Profit_by_Segment['Sales_to_Profit_ratio'] = Sales_and_Profit_by_Segment['Sales']/Sales_and_Profit_by_Segment['Profit']
print(Sales_and_Profit_by_Segment[['Segment','Sales_to_Profit_ratio']])
```
